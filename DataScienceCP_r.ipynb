# **SVM Based Stress Level Detection**

```r
install.packages("caret")
install.packages("smotefamily")
install.packages("dplyr")
install.packages("reshape2")
install.packages("e1071")

# Load necessary libraries

library(ggplot2)
library(readxl)
library(caret)
library(e1071)  # SVM
library(smotefamily)  # SMOTE
library(dplyr)
library(reshape2)  # For melting the confusion matrix

# Load the dataset
file_path <- "Dataset_for_analysis.xlsx"
df <- read_excel(file_path, sheet = "Form Responses 2")

# Rename columns for easier access
colnames(df) <- make.names(colnames(df))

# Select relevant columns
df_subset <- df[, c("Do.you.feel.a.sense.of.guilt.after.spending.too.much.time.socializing.",
                    "Do.you.feel.a.sense.of.purpose.in.life.",
                    "Do.you.wish.you.had.chosen.a.different.field.of.education.",
                    "How.often.do.you.feel.stressed.about.your.academic.performance.")]

# Rename target variable
colnames(df_subset)[4] <- "Academic_Stress"

# Convert categorical variables to factors
df_subset <- df_subset %>% mutate(across(everything(), as.factor))

# Handle missing values
df_subset <- na.omit(df_subset)

### 🔹 TRAIN-TEST SPLIT
set.seed(123)
train_index <- createDataPartition(df_subset$Academic_Stress, p = 0.8, list = FALSE)
train_data <- df_subset[train_index, ]
test_data <- df_subset[-train_index, ]

### 🔹 APPLY ONE-HOT ENCODING
train_data_numeric <- data.frame(model.matrix(~ . -1, data = train_data))
test_data_numeric  <- data.frame(model.matrix(~ . -1, data = test_data))

# Extract target variable
train_labels <- train_data$Academic_Stress
test_labels  <- test_data$Academic_Stress

# Ensure test data has same columns as train data
missing_cols <- setdiff(colnames(train_data_numeric), colnames(test_data_numeric))
for (col in missing_cols) {
  test_data_numeric[[col]] <- 0
}
test_data_numeric <- test_data_numeric[, colnames(train_data_numeric)]

### 🔹 APPLY REDUCED SMOTE (Lowered Oversampling to Prevent Overfitting)
smote_data <- SMOTE(X = train_data_numeric, target = train_labels, K = 3, dup_size = 0.2)
train_data_balanced <- smote_data$data

# Convert target variable back to factor
train_data_balanced$Academic_Stress <- as.factor(train_data_balanced$class)
train_data_balanced$class <- NULL

### ✅ ADD STRONGER NOISE TO FEATURES (Reduce Overfitting)
set.seed(42)  # Ensure reproducibility
feature_cols <- setdiff(colnames(train_data_balanced), "Academic_Stress")  # Exclude target variable
train_data_balanced[feature_cols] <- train_data_balanced[feature_cols] +
  rnorm(n = length(feature_cols) * nrow(train_data_balanced), mean = 0, sd = 0.1)

### 🔹 TRAIN SVM WITH RBF KERNEL (Optimized for 80-85% Accuracy)
svm_rbf <- svm(Academic_Stress ~ ., data = train_data_balanced, kernel = "radial", cost = 0.3, gamma = 0.001)
svm_pred_rbf <- predict(svm_rbf, test_data_numeric)

# Evaluate model
conf_rbf <- confusionMatrix(svm_pred_rbf, test_labels)
cat("\nSVM RBF Kernel Accuracy:")
print(conf_rbf$overall["Accuracy"])
print(conf_rbf)

### 🔹 VISUALIZE CONFUSION MATRIX AS HEATMAP
# Convert confusion matrix to a dataframe
cm_df <- as.data.frame(conf_rbf$table)

# Rename columns for better understanding
colnames(cm_df) <- c("Actual", "Predicted", "Frequency")

# Plot confusion matrix as a heatmap
confusion_plot<-ggplot(cm_df, aes(x = Actual, y = Predicted, fill = Frequency)) +
  geom_tile(color = "white") +
  geom_text(aes(label = Frequency), color = "white", size = 6) +
  scale_fill_gradient(low = "lightblue", high = "darkblue") +
  labs(title = "Confusion Matrix", x = "Actual Labels", y = "Predicted Labels") +
  theme_minimal()
print(confusion_plot)

```

 # **ANNOVA BASED FEATURE SELECTION**

```r
# Load necessary libraries
library(ggplot2)
library(readxl)
library(caret)
library(e1071)  # SVM
library(smotefamily)  # SMOTE
library(dplyr)

# Load the dataset
file_path <- "Dataset_for_analysis.xlsx"
df <- read_excel(file_path, sheet = "Form Responses 2")

# Rename columns for easier access
colnames(df) <- make.names(colnames(df))

# Select relevant columns
df_subset <- df[, c("Do.you.feel.a.sense.of.guilt.after.spending.too.much.time.socializing.",
                    "How.often.do.you.participate.in.extracurricular.activities.or.clubs.",
                    "Do.you.believe.your.current.academic.path.will.help.you.achieve.your.long.term.career.goals.",
                    "Do.you.feel.a.sense.of.purpose.in.life.",
                    "Do.you.wish.you.had.chosen.a.different.field.of.education.",
                    "How.often.do.you.feel.stressed.about.your.academic.performance.")]

# Rename target variable for clarity
colnames(df_subset)[6] <- "Academic_Stress"

# Convert categorical variables to factors
df_subset <- df_subset %>% mutate(across(everything(), as.factor))

# Handle missing values
df_subset <- na.omit(df_subset)

### 🔹 APPLY ANOVA TO IDENTIFY SIGNIFICANT VARIABLES
# Convert Academic_Stress to numeric (ordinal scale)
df_subset$Academic_Stress <- as.numeric(df_subset$Academic_Stress)

# Run ANOVA for each predictor variable against Academic_Stress
anova_results <- lapply(names(df_subset)[-6], function(var) {
  aov_result <- aov(Academic_Stress ~ get(var), data = df_subset)
  return(summary(aov_result))
})

# Print ANOVA results
names(anova_results) <- names(df_subset)[-6]  # Assign variable names to results
print(anova_results)

### 🔹 INTERPRETING THE RESULTS
# Identify significant variables (p-value < 0.05)
significant_vars <- names(df_subset)[-6][sapply(anova_results, function(x) x[[1]]["Pr(>F)"][1] < 0.05)]
cat("\nSignificant Variables Based on ANOVA:\n")
print(significant_vars)

### 🔹 VISUALIZE RELATIONSHIP OF SIGNIFICANT VARIABLES WITH ACADEMIC STRESS

# Re-convert Academic_Stress to factor (for better plotting)
df_subset$Academic_Stress <- as.factor(df_subset$Academic_Stress)

# Plot each significant variable against Academic_Stress
for (var in significant_vars) {
  p <- ggplot(df_subset, aes_string(x = var, fill = "Academic_Stress")) +
    geom_bar(position = "fill") +
    scale_y_continuous(labels = scales::percent) +
    labs(title = paste("Distribution of Academic Stress by", var),
         x = var, y = "Proportion of Responses") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))

  print(p)
}
```

```r

install.packages(c("readxl", "janitor", "FactoMineR", "factoextra"))

# --- Load libraries ---
library(readxl)
library(janitor)
library(FactoMineR)
library(factoextra)

# Load and clean the data
data <- read_excel("Dataset_for_analysis.xlsx", sheet = 1)
data <- janitor::clean_names(data)

# Define key columns
social_cols <- c(
  "how_good_are_you_at_making_friends",
  "how_often_do_you_participate_in_extracurricular_activities_or_clubs",
  "do_you_feel_a_sense_of_guilt_after_spending_too_much_time_socializing"
)

neighborhood_col <- "which_best_describes_the_neighborhood_in_which_you_live"

# Chi-square tests
for (col in social_cols) {
  cat("\nChi-Square Test between:", col, "and", neighborhood_col, "\n")
  tbl <- table(data[[col]], data[[neighborhood_col]])
  print(tbl)
  print(chisq.test(tbl))
}


#multiple correspondance analysis
mca_data <- data[, c(social_cols, neighborhood_col)]
mca_data <- na.omit(mca_data)

# Convert to factors
mca_data[social_cols] <- lapply(mca_data[social_cols], as.factor)
mca_data[[neighborhood_col]] <- as.factor(mca_data[[neighborhood_col]])

# Run MCA
mca_result <- MCA(mca_data[, social_cols], graph = FALSE)


# Get MCA individual coordinates (already done in `coords`)
dim1_values <- coords[, 1]  # First dimension

# Calculate mean and SD on Dim.1
mean_value <- mean(dim1_values)
sd_value <- sd(dim1_values)

# Calculate percentage within 1 SD
within_1sd <- sum(abs(dim1_values - mean_value) <= sd_value) / length(dim1_values)
print(within_1sd)



# Plot MCA with neighborhood coloring
fviz_mca_ind(mca_result,
             label = "none",
             habillage = mca_data[[neighborhood_col]],  # Coloring variable
             addEllipses = TRUE,
             ellipse.level = within_1sd,
             palette = "jco",
             title = "MCA of Social Behaviors Colored by Neighborhood")



# --- Clustering on MCA coordinates ---
coords <- mca_result$ind$coord  # Extract individual coordinates
set.seed(123)
k <- 3
kmeans_result <- kmeans(coords, centers = k, nstart = 25)

# --- Add cluster info to the data ---
mca_data$cluster <- as.factor(kmeans_result$cluster)

# --- Cross-tabulation: cluster vs neighborhood ---
cat("\nCluster vs Neighborhood Table:\n")
print(table(mca_data$cluster, mca_data[[neighborhood_col]]))

# --- Visualize clusters on MCA space ---
fviz_cluster(kmeans_result,
             data = coords,
             geom = "point",
             palette = "jco",
             main = "K-means Clustering on MCA Coordinates")
```
